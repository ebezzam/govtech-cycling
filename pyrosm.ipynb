```python
# https://pyrosm.readthedocs.io/en/latest/basics.html#read-boundaries
h = osm.get_boundaries(boundary_type="administrative", name="Thurgau")
```

```python
# coordinates of the bounding box
bbox_geom = h['geometry'].values[0]
# print(bbox_geom)
```

```python
osm_canton = OSM(fp, bounding_box=bbox_geom)
nodes, edges = osm_canton.get_network(nodes=True, network_type=network_type)
```
